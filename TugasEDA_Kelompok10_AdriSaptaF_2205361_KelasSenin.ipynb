pada dataset jumlah fasilitas kesehatan 2022 untuk kolom jumlah klinik pratama dan jumlah posyandu tidak terisi sama sekali seluruhnya, namun pada dataset 2021 dan 2023 kolom tersebut mengandung value, untuk penanganan mungkin bisa antara dengan mengisi value dengan 0 namun dirasa tidak masuk akal karena di tahun 2021 dan 2023 fasilitas tersebut ada, langkah kedua bisa mengisi dengan sembarang value namun dikhwatirkan data tidak akan valid, maka oleh karena itu saya melakukan drop pada 2 kolom tersebut lalu melakukan hal yang sama terhadap kolom tersebut pada dataset jumlah fasilitas rumah sakit di 2021 dan 2023

```python
# Menghapus kolom 'Jumlah Klinik Pratama' dan 'Jumlah Posyandu' dari df_rs21, df_rs22, dan df_rs23
df_rs21.drop(['Jumlah Klinik Pratama', 'Jumlah Posyandu'], axis=1, inplace=True)
df_rs22.drop(['Jumlah Klinik Pratama', 'Jumlah Posyandu'], axis=1, inplace=True)
df_rs23.drop(['Jumlah Klinik Pratama', 'Jumlah Posyandu'], axis=1, inplace=True)

# Tampilkan dataframe setelah penghapusan kolom untuk verifikasi
print("df_rs21 setelah penghapusan kolom:\n", df_rs21.head(), "\n")
print("df_rs22 setelah penghapusan kolom:\n", df_rs22.head(), "\n")
print("df_rs23 setelah penghapusan kolom:\n", df_rs23.head(), "\n")
```

sekarang masih terdapat beberapa missing value namun berbeda dengan 2 kolom sebelumnya yang full kosong untuk penanganan nilai missing value ini saya mengisi nilai kosong tersebut dengan median karena mengisi nilai dengan median dianggap lebih stabil jika terdapat nilai ekstrim

```python
# Mengisi missing values (NaN) pada df_rs21, df_rs22, dan df_rs23 dengan nilai median masing-masing kolom
df_rs21.fillna(df_rs21.median(numeric_only=True), inplace=True)
df_rs22.fillna(df_rs22.median(numeric_only=True), inplace=True)
df_rs23.fillna(df_rs23.median(numeric_only=True), inplace=True)

# Tampilkan dataframe setelah pengisian nilai NaN untuk verifikasi
print("df_rs21 setelah pengisian nilai NaN:\n", df_rs21.head(), "\n")
print("df_rs22 setelah pengisian nilai NaN:\n", df_rs22.head(), "\n")
print("df_rs23 setelah pengisian nilai NaN:\n", df_rs23.head(), "\n")
```

setelah diisi sekarang dapat dilihat bahwa tipe data pada setiap kolom menggunakan desimal, karena itu saya mengubah semua tipe data ke integer karena saya rasa jumlah rumah sakit pasti bilangan bulat tidak ada jumlah rumah sakit bernilai desimal misalnya jumlah rumah sakit umum di suatu daerah totalnya da 2.5 menurut saya sangat tidak masuk akal sehingga saya ubah semua ke tipe bilagan bulat

```python
# Ubah semua kolom yang bertipe float menjadi int pada df_rs21
df_rs21 = df_rs21.astype({col: 'int' for col in df_rs21.select_dtypes(include=['float']).columns})

# Ubah semua kolom yang bertipe float menjadi int pada df_rs22
df_rs22 = df_rs22.astype({col: 'int' for col in df_rs22.select_dtypes(include=['float']).columns})

# Ubah semua kolom yang bertipe float menjadi int pada df_rs23
df_rs23 = df_rs23.astype({col: 'int' for col in df_rs23.select_dtypes(include=['float']).columns})

# Tampilkan tipe data dari ketiga dataframe setelah perubahan
print("Tipe data df_rs21 setelah diubah:\n", df_rs21.dtypes, "\n")
print("Tipe data df_rs22 setelah diubah:\n", df_rs22.dtypes, "\n")
print("Tipe data df_rs23 setelah diubah:\n", df_rs23.dtypes, "\n")
```

15. Merge Dataset berdasarkan tahun

```python
# 1. Menghitung jumlah total fasilitas rumah sakit per tahun
total_fasilitas_rumah_sakit_provinsi = [
    df_hospitals['Jumlah Rumah Sakit Umum_' + str(year)].sum() +
    df_hospitals['Jumlah Rumah Sakit Khusus_' + str(year)].sum() +
    df_hospitals['Jumlah Puskesmas Rawat Inap_' + str(year)].sum() +
    df_hospitals['Jumlah Puskesmas Non Rawat Inap_' + str(year)].sum()
    for year in [2021, 2022, 2023]
]

# 2. Menghitung rata-rata usia harapan hidup dari laki-laki dan perempuan per tahun
rata_rata_uhh_per_tahun = [
    df_UHH2023[['Laki-Laki_' + str(year), 'Perempuan_' + str(year)]].mean().mean()
    for year in [2021, 2022, 2023]
]

# 3. Menghitung rata-rata persentase perokok per tahun
rata_rata_perokok_per_tahun = [
    df_perokok_combined[[f'Merokok_15_24_{year}', f'Merokok_25_34_{year}',
                         f'Merokok_35_44_{year}', f'Merokok_45_54_{year}',
                         f'Merokok_55_64_{year}', f'Merokok_65_plus_{year}']].mean().mean()
    for year in [2021, 2022, 2023]
]

# 4. Membuat dataframe untuk provinsi dengan data per tahun
df_provinsi = pd.DataFrame({
    'Tahun': ['2021', '2022', '2023'],
    'Total Fasilitas Rumah Sakit': total_fasilitas_rumah_sakit_provinsi,
    'Rata-rata Usia Harapan Hidup': rata_rata_uhh_per_tahun,
    'Rata-rata Persentase Perokok': rata_rata_perokok_per_tahun
})

# 5. Visualisasi menggunakan lineplot dengan sumbu y sekunder
fig, ax1 = plt.subplots(figsize=(14, 7))

# Plot untuk Total Fasilitas Rumah Sakit
ax1.set_xlabel('Tahun')
ax1.set_ylabel('Total Fasilitas Rumah Sakit', color='blue')
sns.lineplot(data=df_provinsi, x='Tahun', y='Total Fasilitas Rumah Sakit', marker='o', label='Total Fasilitas Rumah Sakit', color='blue', ax=ax1)
ax1.tick_params(axis='y', labelcolor='blue')

# Sumbu y sekunder untuk Rata-rata Usia Harapan Hidup
ax2 = ax1.twinx()
ax2.set_ylabel('Rata-rata Usia Harapan Hidup', color='orange')
sns.lineplot(data=df_provinsi, x='Tahun', y='Rata-rata Usia Harapan Hidup', marker='o', label='Rata-rata Usia Harapan Hidup', color='orange', ax=ax2)
ax2.tick_params(axis='y', labelcolor='orange')

# Sumbu y sekunder untuk Rata-rata Persentase Perokok
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))  # Menggeser posisi sumbu y ketiga
ax3.set_ylabel('Rata-rata Persentase Perokok', color='green')
sns.lineplot(data=df_provinsi, x='Tahun', y='Rata-rata Persentase Perokok', marker='o', label='Rata-rata Persentase Perokok', color='green', ax=ax3)
ax3.tick_params(axis='y', labelcolor='green')

plt.title('Hubungan antara Total Fasilitas Rumah Sakit, Rata-rata Usia Harapan Hidup, dan Rata-rata Persentase Perokok di Jawa Barat (2021-2023)')
fig.tight_layout()
plt.grid()
plt.show()
```

Dari 2021 hingga 2022, persentase perokok meningkat tajam, sementara rata-rata usia harapan hidup juga meningkat, tetapi dalam laju yang lebih lambat. Namun, ketika persentase perokok menurun sedikit di 2023, kenaikan usia harapan hidup masih berlanjut. Ini bisa menjadi indikasi bahwa kebiasaan merokok mungkin belum menunjukkan dampak langsung dalam jangka pendek terhadap usia harapan hidup pada rentang waktu yang singkat (hanya tiga tahun). Namun, dalam jangka panjang, kebiasaan merokok biasanya memiliki dampak negatif yang signifikan terhadap kesehatan.
